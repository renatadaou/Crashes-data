## 1) Exploring the crashes 

Just by looking at the spreadsheet we can see that there are 136 idependent crahes.

I want to look at juist the Town Name, Date Of Crash, Number Of Motor Vehicles, Light Condition Text Format

```python
import pandas as pd
```

```python
# Load the newly provided CSV file
file_path = 'export_182420_0.csv'
data = pd.read_csv(file_path)

# Display the first few rows to identify the correct columns
data.head()
```

```python
# Extract the relevant columns
filtered_data = data[['Town Name','Latitude','Longitude', 'Date Of Crash', 'Number Of Motor Vehicles', 'Light Condition Text Format']]

# Save the filtered data to a new CSV file
output_file_path = 'filtered_crash_data_final.csv'
filtered_data.to_csv(output_file_path, index=False)

output_file_path
```

```python
#show the data
filtered_data = pd.read_csv('filtered_crash_data_final.csv')
filtered_data
```

Number of fatal crashes per town

```python
# Group the data by 'Town Name' and count the number of crashes for each town
crash_counts_by_town = data.groupby('Town Name').size().reset_index(name='Crash Count')

# Sort the results by 'Crash Count' in descending order
crash_counts_by_town_sorted = crash_counts_by_town.sort_values(by='Crash Count', ascending=False)

```

```python
crash_counts_by_town_sorted
```

Now, surprisingly, most of the crashes happen during the day

```python
# Group the data by 'Light Condition Text' and count the number of instances for each condition
light_condition_counts = data.groupby('Light Condition Text Format').size().reset_index(name='Instance Count')

# Sort the results by 'Instance Count' in descending order
light_condition_counts_sorted = light_condition_counts.sort_values(by='Instance Count', ascending=False)
light_condition_counts_sorted
```
