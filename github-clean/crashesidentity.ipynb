## 2) Filtering the data
Now we only have the information of the people who were killed in the crashes.

I am going to see how many were drivers, how many were passengers, how many were pedestrians. I will also filter to see the most common age of those who were killed and the gender.

```python
# Load the newly uploaded dataset to ensure accuracy
fatal_injury_k_data_path = 'fatal_injury_k_data.csv'
fatal_injury_k_data = pd.read_csv(fatal_injury_k_data_path)

# Count the number of drivers, passengers, and pedestrians
driver_count = fatal_injury_k_data[fatal_injury_k_data['Person Type Text Format'] == 'Driver'].shape[0]
passenger_count = fatal_injury_k_data[fatal_injury_k_data['Person Type Text Format'] == 'Passenger'].shape[0]
pedestrian_count = fatal_injury_k_data[fatal_injury_k_data['Person Type Text Format'] == 'Pedestrian'].shape[0]

# Most common Age
most_common_age = fatal_injury_k_data['Age'].mode()[0]

# Most common Gender Text
most_common_gender_text = fatal_injury_k_data['Gender Text Format'].mode()[0]

driver_count, passenger_count, pedestrian_count, most_common_age, most_common_gender_text
```

I also want to see how many people were killed in DUI accidents (Under the Influence of Medications/Drugs/Alcohol)

```python
# Filter the data where "Condition at Time of Crash Text Format" is "Under the Influence of Medications/Drugs/Alcohol"
under_influence_data = fatal_injury_k_data[fatal_injury_k_data['Condition at Time of Crash Text Format'] == 'Under the Influence of Medications/Drugs/Alcohol']

# Display the filtered data
under_influence_data
```

Out of curiosity, I wanted to know the most affected age range (5-year age range)

```python
# Create five-year age ranges and count the occurrences in the dataset
fatal_injury_k_data['Age Range'] = pd.cut(fatal_injury_k_data['Age'], bins=range(0, 105, 5))

# Count the number of occurrences in each age range
age_range_counts = fatal_injury_k_data['Age Range'].value_counts().sort_index()

# Find the most affected age range
most_affected_age_range = age_range_counts.idxmax()
most_affected_age_range_count = age_range_counts.max()

most_affected_age_range, most_affected_age_range_count
```
